 **importing libraries**

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Initialize an empty list to store all elements
all_elements = []

# Loop through each page and fetch the content
for i in range(81):
    page = requests.get(f"https://arab-books.com/page/{i}/").text
    soup = BeautifulSoup(page, 'lxml')
    elements = soup.find_all('div', class_="col-lg-3 col-md-4 col-6")
    all_elements.extend(elements)  # Add elements to the list
    print(f"Processing page {i}, found {len(elements)} elements")

# Check the total number of elements
print(f"Total elements found: {len(all_elements)}")
```

```python
import re
url_pattern = "(https://arab-books.com/books/([^\"]+)/)"
URl_RegEx = re.compile(url_pattern)

Books_urls=[]
for book in all_elements :
  myResult = URl_RegEx.search(str(book))
  Books_urls.append(myResult.group())
Books_urls[:]
```

```python
all_details = []

for i in range(len(Books_urls)):
  book = requests.get(Books_urls[i]).text
  Beautiful_book = BeautifulSoup(book, 'lxml')
  details = Beautiful_book.find_all('li',class_="mb-2")
  all_details.extend(details)
  print(f"Processing book {i}, found {len(details)} details")

# Check the total number of elements
print(f"Total details found: {len(all_details)}")
```

```python
# Initialize an empty list to store the processed details
processed_list_of_lists = []

for i in range(len(Books_urls)):
    book = requests.get(Books_urls[i]).text
    Beautiful_book = BeautifulSoup(book, 'lxml')
    details = Beautiful_book.find_all('li', class_="mb-2")

    # Extract text content from each detail and store it in a sublist
    sublist = [detail.get_text(strip=True) for detail in details]
    processed_list_of_lists.append(sublist)

# Check the total number of sublists (individual details)
total_sublists = len(processed_list_of_lists)
print(f"Total sublists (individual details) found: {total_sublists}")

# Example: Print the first sublist (details from the first book)
if total_sublists > 0:
    print("Example sublist (details from the first book):")
    print(processed_list_of_lists[0])
```
