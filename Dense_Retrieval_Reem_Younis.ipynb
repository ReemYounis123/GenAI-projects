```python
from sentence_transformers import SentenceTransformer

# Load a pre-trained model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Create embeddings for each chunk
embeddings = model.encode(chunks)

print(embeddings.shape)  # Check the shape of the embeddings
```

```python
import numpy as np

# Save embeddings to a file
np.save('book_embeddings.npy', embeddings)

# Download the file
from google.colab import files
files.download('book_embeddings.npy')
```

```python
query = "What is ROS"  # Replace with your actual query
query_embedding = model.encode(query)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Calculate cosine similarity between the query and each chunk
cosine_similarities = cosine_similarity(
    [query_embedding], embeddings)[0]

# Find the index of the most similar chunk
most_similar_idx = np.argmax(cosine_similarities)

# Retrieve the most similar chunk
most_similar_chunk = chunks[most_similar_idx]

print(f"Most similar chunk:\n{most_similar_chunk}")
```

```python
def find_similar_chunk(query, similarity_threshold=0.7):
    # Create an embedding for the query
    query_embedding = model.encode(query)

    # Calculate cosine similarity between the query and each chunk
    cosine_similarities = cosine_similarity([query_embedding], embeddings)[0]

    # Find the index of the most similar chunk
    most_similar_idx = np.argmax(cosine_similarities)

    # Retrieve the most similar chunk and its similarity score
    most_similar_chunk = chunks[most_similar_idx]
    similarity_score = cosine_similarities[most_similar_idx]

    # Check if the similarity score is above the threshold
    if similarity_score >= similarity_threshold:
        return f"Most similar chunk:\n\n{most_similar_chunk}\n\nSimilarity: {similarity_score:.4f}"
    else:
        return "No relevant chunk found with sufficient similarity."

```

```python

import gradio as gr
# Set up Gradio interface
Dense_retrieval = gr.Interface(
    fn=find_similar_chunk,
    inputs="text",
    outputs="text",
    title="Book Query Search",
    description="Enter a query, and the application will find the most similar chunk from the book."
)

# Launch the interface
Dense_retrieval.launch()
```
